## 2. What kind of API did you build? What types of APIs are there? Why are APIs important?

#### What is API?

* API(Application Programming Interface) is a messenger between server and us.It delivers our requests to provider that we are requesting it and come back to us with their responses. API communicates from one software to another. 

#### When to create APIs?

* If our dataset is large, if we need to access one part of data
* If data changes or updated frequently
* If we want to performs actions other than retriving data sunch as updating and deleting

These are some of the reasons to create APIs.

#### What types of APIs are there?

* web based API
* database based API
* OS based API
* hardware based API

#### What kind of API did you build?

* I have created **WEB API**- an API which can be accessed over web using HTTP based services.

#### Why are APIs important?

* APIs are vital tools for businesses in all industries. The importance of APIs are, they allow the capabilities of one computer program to be used by another. So that different programs are able to communicate.
